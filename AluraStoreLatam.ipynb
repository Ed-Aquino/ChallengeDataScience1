# 3. Calificación promedio de la tienda


```python
# Funcion para conocer la media de calificacion para cada tienda

def calificacion(tienda):
  mediaCalificaciones = 0
  totalCalificaciones = 0
  for calificacion in tienda['Calificación']:
    int(calificacion)
    mediaCalificaciones += calificacion
    totalCalificaciones += 1
  media = round(mediaCalificaciones/ totalCalificaciones, 2)
  return media
```

```python
# Valoracion promedio por cada tienda

def valoracion(mediaTienda):
  if mediaTienda >= 4.0:
    print(f'Valoracion excelente: {mediaTienda}')
  elif mediaTienda >= 3.0:
    print(f'Valoracion buena: {mediaTienda}')
  elif mediaTienda >= 2.0:
    print(f'Valoracion intermedia: {mediaTienda}')
  elif mediaTienda >= 1.0:
    print(f'Valoracion baja: {mediaTienda}')
  else:
    print(f'Valoracion mala: {mediaTienda}')
```

```python
# Resultados para cada tienda
print("TIENDA 1")
valoracion(calificacion(tienda1))
print("\nTIENDA 2")
valoracion(calificacion(tienda2))
print("\nTIENDA 3")
valoracion(calificacion(tienda3))
print("\nTIENDA 4")
valoracion(calificacion(tienda4))
```

# 4. Productos más y menos vendidos

```python
# funcion para contar los productos

def productosVendidos(tienda):
  cantidad = 0
  productosCantidad = []
  productosTienda = tienda['Producto']
  productos = []
  for producto in productosTienda:
    if producto in productos:
      pass
    else:
      productos.append(producto)
  for producto in productos:
    cantidad = int(productosTienda[productosTienda == producto].count())
    productosCantidad.append({'Producto': producto , 'Cantidad': cantidad})
    productosOrdenados = sorted(productosCantidad, key=lambda d: d['Cantidad'], reverse=True)
  #productosMasVendidos = productosOrdenados[:3]
  return productosOrdenados
```

```python
print(f"Productos ordenados por la cantidad de ventas para tienda 1 es: \n{productosVendidos(tienda1)}")
print(f"\Productos ordenados por la cantidad de ventas para tienda 2 es: \n{productosVendidos(tienda2)}")
print(f"\Productos ordenados por la cantidad de ventas para tienda 3 es: \n{productosVendidos(tienda3)}")
print(f"\Productos ordenados por la cantidad de ventas para tienda 4 es: \n{productosVendidos(tienda4)}")
```

# 5. Envío promedio por tienda

```python
# Funcion para calcular el costo promedio para cada tienda

def costoEnvio(tienda):
  costoTotal = 0.0
  contador = 0.0
  for costo in tienda['Costo de envío']:
    costoTotal += float(costo)
    contador += 1.0
  promedioEnvio = round(costoTotal/contador, 2 )
  return promedioEnvio
```

```python
# Obteniendo el promedio del costo de envio para cada tienda

print(f"El costo promedio de envio para la tienda 1 es de {costoEnvio(tienda1)}")
print(f"El costo promedio de envio para la tienda 2 es de {costoEnvio(tienda2)}")
print(f"El costo promedio de envio para la tienda 3 es de {costoEnvio(tienda3)}")
print(f"El costo promedio de envio para la tienda 4 es de {costoEnvio(tienda4)}")
```
